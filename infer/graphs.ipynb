```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

size = np.arange(1, 100)
extracted_times = 1e3 * np.loadtxt('extracted.csv', delimiter=',', skiprows=2, usecols=1)
map_times =       1e3 * np.loadtxt('map.csv',       delimiter=',', skiprows=2, usecols=1)

a, b, c = np.polyfit(size, extracted_times, 2)
extracted_fit = lambda x: a * x ** 2 + b * x + c

a, b, c = np.polyfit(size, map_times, 2)
map_fit = lambda x: a * x ** 2 + b * x + c

plt.figure(figsize=(12, 7))
plt.title("extracted code performance")
plt.xlabel("input size (nodes)")
plt.ylabel("processor time (ms)")
plt.scatter(size, extracted_times, color='black', marker='+')
plt.plot(size, extracted_fit(size), color='black')
plt.savefig('extracted.pdf')
plt.show()

plt.figure(figsize=(12, 7))
plt.title("Data.Map performance")
plt.xlabel("input size (nodes)")
plt.ylabel("processor time (ms)")
plt.scatter(size, map_times, color='black', marker='+')
plt.plot(size, map_fit(size), color='black')
plt.savefig('map.pdf')
plt.show()
```
