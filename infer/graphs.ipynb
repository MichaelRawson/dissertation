```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

FIG_SIZE = (12, 7)

def load_file(name):
    file = 'bench/' + name + '.csv'
    size = np.array(
        [int(f.split('/')[1]) for f in np.genfromtxt(file, delimiter=',', skip_header=1, usecols=0, dtype=None)]
    )
    time = np.genfromtxt(file, delimiter=',', skip_header=1, usecols=1)
    return (size, time)

def produce_performance_graph(name):
    size, time = load_file(name)

    a, b, r, _, _ = stats.linregress(np.log(size), np.log(time))
    fitted = np.exp(b) * size ** a

    plt.figure(figsize=FIG_SIZE)
    plt.xscale('log')
    plt.yscale('log')
    plt.scatter(size, time, color='black', marker='x', label='benchmark data')
    plt.plot(size, fitted, color='black', linestyle=':', label='regression, $r^2 = %.3f$' % r ** 2)

    plt.title("asymptotic performance")
    plt.xlabel("input size (nodes)")
    plt.ylabel("processor time (s)")
    plt.legend()

    plt.savefig('plots/' + name + '.pdf')
    plt.show()
```

```python
produce_performance_graph('welltyped')
```
